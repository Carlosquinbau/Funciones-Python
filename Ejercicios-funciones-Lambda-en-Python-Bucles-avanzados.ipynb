# Ejercicio 2
Escriba una función de Python para dar la suma de los números de una lista.

````python
lista=[2,3,4,5,6,7]

def suma(lista):                 #Función que realiza la suma de los números de lista.
    contador=0
    for numero in lista:         #Bucle for que reccore la lista
        contador=contador+numero #Contador que va sumando los numeros de la lista.
    return contador

print(suma(lista))
````

#Ejercicio 9

Dada una palabra, cuente el número de vocales que tiene.



````python
def contar_vocales(palabra):
    vocales=["a", "e", "i", "o", "u"]
    contador=0
    for vocal in vocales:
        if vocal in palabra:
            contador +=1
    return contador

print(contar_vocales("Hablemos"))


````
